```python
def run_excel(path):
    pathes = os.listdir(path)
    for _path in pathes:
        _path = f'{path}/{_path}'
        if os.path.isfile(_path): 
            if _path.split('.')[-1] == 'xlsx':
                print(_path)
        else:
            run_excel(_path)
```
